```python
import os
import glob
import random
import rasterio
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
import timm
from torch.utils.data import Dataset, DataLoader
from sklearn.model_selection import train_test_split
from torchvision.transforms import functional as TF

# ‚úÖ Paths
BASE_DIR = "/content/drive/MyDrive/"
PRE_EVENT_DIR = os.path.join(BASE_DIR, "sar_pre_event_images")
POST_EVENT_DIR = os.path.join(BASE_DIR, "sa_post_event_images")
PRE_MASK_DIR = os.path.join(BASE_DIR, "pre_mask")
POST_MASK_DIR = os.path.join(BASE_DIR, "post_mask")
MODEL_PATH = os.path.join(BASE_DIR, "best_model.pth")

class LandslideDataset(Dataset):
    def __init__(self, file_ids, pre_dir, post_dir, pre_mask_dir, post_mask_dir, image_size=(224, 224)):
        self.file_ids = file_ids
        self.pre_dir = pre_dir
        self.post_dir = post_dir
        self.pre_mask_dir = pre_mask_dir
        self.post_mask_dir = post_mask_dir
        self.image_size = image_size

    def _load_image(self, path):
        with rasterio.open(path) as src:
            image = src.read(1)
            image = (image - np.min(image)) / (np.max(image) - np.min(image) + 1e-6)
            tensor = torch.tensor(image, dtype=torch.float32).unsqueeze(0)
            tensor = F.interpolate(tensor.unsqueeze(0), size=self.image_size, mode='bilinear', align_corners=False).squeeze(0)
        return tensor

    def _load_mask(self, path):
        with rasterio.open(path) as src:
            mask = src.read(1)
            mask = (mask > 0).astype(np.float32)
            mask = torch.tensor(mask).unsqueeze(0)
            mask = F.interpolate(mask.unsqueeze(0), size=self.image_size, mode='nearest').squeeze(0)
        return mask

    def __len__(self):
        return len(self.file_ids)

    def __getitem__(self, idx):
        file_id = self.file_ids[idx]

        pre_img = self._load_image(os.path.join(self.pre_dir, f"pre_{file_id}.tif"))
        post_img = self._load_image(os.path.join(self.post_dir, f"post_{file_id}.tif"))

        pre_mask = self._load_mask(os.path.join(self.pre_mask_dir, f"pre_mask_{file_id}.tif"))
        post_mask = self._load_mask(os.path.join(self.post_mask_dir, f"post_{file_id}.tif"))

        combined_img = torch.cat((pre_img, post_img), dim=0)
        combined_mask = torch.logical_or(pre_mask, post_mask).float()

        return combined_img, combined_mask

def debug_folder_counts():
    pre_ids = {f.replace("pre_", "").replace(".tif", "") for f in os.listdir(PRE_EVENT_DIR)}
    post_ids = {f.replace("post_", "").replace(".tif", "") for f in os.listdir(POST_EVENT_DIR)}
    pre_mask_ids = {f.replace("pre_mask_", "").replace(".tif", "") for f in os.listdir(PRE_MASK_DIR)}
    post_mask_ids = {f.replace("post_", "").replace(".tif", "") for f in os.listdir(POST_MASK_DIR)}

    print(f"\U0001F4C1 Pre Images: {len(pre_ids)}")
    print(f"\U0001F4C1 Post Images: {len(post_ids)}")
    print(f"\U0001F4C1 Pre Masks: {len(pre_mask_ids)}")
    print(f"\U0001F4C1 Post Masks: {len(post_mask_ids)}")

    valid_ids = pre_ids & post_ids & pre_mask_ids & post_mask_ids
    print(f"‚úÖ Valid Pairs (intersection): {len(valid_ids)}")

def get_file_ids():
    pre_ids = {f.replace("pre_", "").replace(".tif", "") for f in os.listdir(PRE_EVENT_DIR)}
    post_ids = {f.replace("post_", "").replace(".tif", "") for f in os.listdir(POST_EVENT_DIR)}
    pre_mask_ids = {f.replace("pre_mask_", "").replace(".tif", "") for f in os.listdir(PRE_MASK_DIR)}
    post_mask_ids = {f.replace("post_", "").replace(".tif", "") for f in os.listdir(POST_MASK_DIR)}
    valid_ids = pre_ids & post_ids & pre_mask_ids & post_mask_ids
    return sorted(list(valid_ids))

debug_folder_counts()

# ‚úÖ Split data
file_ids = get_file_ids()
print(f"üî¢ Total Paired Image IDs: {len(file_ids)}")

train_ids, valtest_ids = train_test_split(file_ids, test_size=0.3, random_state=42)
val_ids, test_ids = train_test_split(valtest_ids, test_size=0.5, random_state=42)

train_loader = DataLoader(LandslideDataset(train_ids, PRE_EVENT_DIR, POST_EVENT_DIR, PRE_MASK_DIR, POST_MASK_DIR),
                          batch_size=8, shuffle=True, num_workers=0)
val_loader = DataLoader(LandslideDataset(val_ids, PRE_EVENT_DIR, POST_EVENT_DIR, PRE_MASK_DIR, POST_MASK_DIR),
                        batch_size=8, shuffle=False, num_workers=0)

print(f"‚úÖ Valid Image Triplets (Pre+Post+Mask): {len(file_ids)}")
print(f"üîπ Train: {len(train_ids)}, Val: {len(val_ids)}, Test: {len(test_ids)}")

# ‚úÖ Model
class SegmentationModel(nn.Module):
    def __init__(self, num_classes=1):
        super().__init__()
        self.channel_expand = nn.Conv2d(2, 3, kernel_size=1)
        self.efficientnet = timm.create_model("efficientnet_b0", pretrained=True, features_only=True)
        self.swin = timm.create_model("swin_tiny_patch4_window7_224", pretrained=True, features_only=True)
        self.projection = nn.Conv2d(1088, 1024, kernel_size=1)
        self.conv1 = nn.Conv2d(1024, 512, kernel_size=3, padding=1)
        self.conv2 = nn.Conv2d(512, num_classes, kernel_size=1)

    def forward(self, x):
        x = self.channel_expand(x)
        eff = self.efficientnet(x)[-1]
        swin = self.swin(x)[-1].permute(0, 3, 1, 2)
        swin = F.interpolate(swin, size=eff.shape[2:], mode="bilinear")
        x = torch.cat([eff, swin], dim=1)
        x = self.projection(x)
        x = F.relu(self.conv1(x))
        x = self.conv2(x)
        x = F.interpolate(x, size=(224, 224), mode="bilinear")
        return x

# ‚úÖ Training Setup
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
model = SegmentationModel().to(DEVICE)
optimizer = torch.optim.Adam(model.parameters(), lr=5e-4, weight_decay=1e-4)
criterion = nn.BCEWithLogitsLoss()
scaler = torch.cuda.amp.GradScaler()
scheduler = torch.optim.lr_scheduler.ReduceLROnPlateau(optimizer, patience=2, factor=0.5, verbose=True)

train_losses, val_losses, val_ious = [], [], []

# ‚úÖ IoU accuracy metric
def compute_iou(preds, masks, threshold=0.5):
    preds = torch.sigmoid(preds) > threshold
    intersection = (preds & (masks > 0.5)).sum().float()
    union = ((preds | (masks > 0.5))).sum().float()
    if union == 0:
        return torch.tensor(1.0)
    return intersection / union

# ‚úÖ Evaluation
def evaluate(model, loader):
    model.eval()
    val_loss, total_iou = 0, 0
    with torch.no_grad():
        for inputs, masks in loader:
            inputs, masks = inputs.to(DEVICE), masks.to(DEVICE)
            with torch.cuda.amp.autocast():
                outputs = model(inputs)
                loss = criterion(outputs, masks)
                val_loss += loss.item()
                total_iou += compute_iou(outputs, masks).item()
    return val_loss / len(loader), total_iou / len(loader)

# ‚úÖ Train Loop with 3 callbacks
def train(model, train_loader, val_loader, epochs=25, patience=5):
    best_loss = float('inf')
    early_stop_counter = 0

    for epoch in range(epochs):
        model.train()
        total_loss = 0

        for inputs, masks in train_loader:
            inputs, masks = inputs.to(DEVICE), masks.to(DEVICE)
            optimizer.zero_grad()

            with torch.cuda.amp.autocast():
                outputs = model(inputs)
                loss = criterion(outputs, masks)

            scaler.scale(loss).backward()
            scaler.step(optimizer)
            scaler.update()

            total_loss += loss.item()

        val_loss, val_iou = evaluate(model, val_loader)
        scheduler.step(val_loss)

        avg_train_loss = total_loss / len(train_loader)
        train_losses.append(avg_train_loss)
        val_losses.append(val_loss)
        val_ious.append(val_iou)

        print(f"‚úÖ Epoch {epoch+1}: Train Loss = {avg_train_loss:.4f} | Val Loss = {val_loss:.4f} | Val IoU = {val_iou:.4f}")
        if val_loss < best_loss:
            best_loss = val_loss
            early_stop_counter = 0
            torch.save(model.state_dict(), MODEL_PATH)
            print("\U0001F4BE Best model saved!")
        else:
            early_stop_counter += 1
            if early_stop_counter >= patience:
                print("‚èπÔ∏è Early stopping triggered!")
                break


train(model, train_loader, val_loader, epochs=25, patience=5)
```

```python
import ee
import requests
import numpy as np
import matplotlib.pyplot as plt
import io
import torch
from datetime import datetime, timedelta
from torchvision import transforms

# Define ROI
roi = ee.Geometry.Polygon([
    [[88.253, 27.031], [88.253, 27.071], [88.303, 27.071], [88.303, 27.031], [88.253, 27.031]]
])

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# üîÅ Load Model (make sure SegmentationModel class is defined)
model = SegmentationModel().to(device)
state_dict = torch.load("/content/drive/MyDrive/best_model.pth", map_location=device)
model.load_state_dict(state_dict)
model.eval()

# üì• Download SAR Image from Earth Engine
def get_sar_image(date):
    collection = ee.ImageCollection('COPERNICUS/S1_GRD') \
        .filterBounds(roi) \
        .filterDate(date, date.advance(3, 'day')) \
        .filter(ee.Filter.eq('instrumentMode', 'IW')) \
        .filter(ee.Filter.listContains('transmitterReceiverPolarisation', 'VV')) \
        .select('VV')

    if collection.size().getInfo() == 0:
        print(f"üö´ No Sentinel-1 image available for: {date.format('YYYY-MM-dd').getInfo()}")
        return None

    try:
        image = collection.median().clip(roi).multiply(0.0001)
        url = image.getDownloadURL({
            'scale': 10,
            'region': roi,
            'format': 'NPY'
        })
        response = requests.get(url)
        array = np.load(io.BytesIO(response.content))
        if array.dtype.names:
            array = array[array.dtype.names[0]]
        return array.astype(np.float32)
    except Exception as e:
        print(f"‚ö†Ô∏è Error while downloading image: {e}")
        return None

# üîÑ Normalize SAR for display
def stretch(img):
    img = img.copy()
    img = img - img.min()
    img = img / (img.max() + 1e-6)
    return img

# üñºÔ∏è Display 4 images: Pre, Post, Mask, Difference
def show_landslide_images(pre_array, post_array, predicted_mask, landslide_percentage):
    difference_image = np.clip(post_array - pre_array, -0.2, 0.2)

    fig, axs = plt.subplots(1, 4, figsize=(16, 4))

    axs[0].imshow(stretch(pre_array), cmap='gray')
    axs[0].set_title('Pre-Event SAR')

    axs[1].imshow(stretch(post_array), cmap='gray')
    axs[1].set_title('Post-Event SAR')

    axs[2].imshow(predicted_mask.squeeze(), cmap='gray')
    axs[2].set_title(f'Landslide Detected:\n{landslide_percentage:.2f}%')

    axs[3].imshow(difference_image, cmap='seismic')
    axs[3].set_title('SAR Difference\n(Post - Pre)')

    for ax in axs:
        ax.axis('off')

    plt.tight_layout()
    plt.show()

# üß™ Preprocess tensors
def preprocess_image(image):
    transform = transforms.Compose([
        transforms.ToTensor(),
        transforms.Resize((224, 224)),
    ])
    return transform(image).unsqueeze(0)

# üö® Detect Landslide
def detect_landslide_on_date(pre_date_str, post_date_str):
    pre_date = ee.Date(pre_date_str)
    post_date = ee.Date(post_date_str)

    shift_attempts = 0
    max_attempts = 7
    pre_array, post_array = None, None

    while shift_attempts < max_attempts:
        print(f"Trying with shifted dates: {pre_date.format('YYYY-MM-dd').getInfo()} & {post_date.format('YYYY-MM-dd').getInfo()}")
        pre_array = get_sar_image(pre_date)
        post_array = get_sar_image(post_date)

        if pre_array is not None and post_array is not None:
            break

        pre_date = pre_date.advance(-1, 'day')
        post_date = post_date.advance(-1, 'day')
        shift_attempts += 1

    if pre_array is None or post_array is None:
        print("‚ö†Ô∏è SAR images not available for selected dates.")
        return

    # Convert to 224x224 tensors
    pre_tensor = preprocess_image(pre_array)
    post_tensor = preprocess_image(post_array)

    input_tensor = torch.cat((pre_tensor, post_tensor), dim=1).to(device)

    with torch.no_grad():
        prediction = model(input_tensor)
        predicted_mask = torch.sigmoid(prediction).squeeze().cpu().numpy()
        binary_mask = (predicted_mask > 0.5).astype(np.uint8)

    landslide_percentage = (np.sum(binary_mask) / binary_mask.size) * 100

    # Resize original arrays to match model input size for consistent visual display
    from skimage.transform import resize
    pre_resized = resize(pre_array, (224, 224), preserve_range=True)
    post_resized = resize(post_array, (224, 224), preserve_range=True)

    show_landslide_images(pre_resized, post_resized, predicted_mask, landslide_percentage)

# üîÅ Example run
detect_landslide_on_date("2023-06-02", "2023-10-04")
```
